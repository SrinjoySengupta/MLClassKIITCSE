```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
X = np.array([1, 5, 3, 4, 5])  
y = np.array([1, 2, 1.3, 3.75, 2.25]) 

X = X.reshape(-1, 1) 

X_b = np.c_[np.ones((X.shape[0], 1)), X]  

theta = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

# 4. Display the parameters (intercept and slope)
print(f"Intercept (theta_0): {theta[0]}")
print(f"Slope (theta_1): {theta[1]}")

# 5. Make predictions based on the learned model
y_pred = X_b.dot(theta)

# 6. Visualize the data and the regression line
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, y_pred, color='red', label='Regression line')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression using Least Squares Method')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 5, 7, 8])

# Calculate the necessary sums
n = len(x)
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_x_squared = np.sum(x**2)
sum_x_y = np.sum(x * y)

# Calculate the coefficients using the least squares formulas
beta_1 = (n * sum_x_y - sum_x * sum_y) / (n * sum_x_squared - sum_x**2)
beta_0 = (sum_y - beta_1 * sum_x) / n

# Print the coefficients
print(f"Intercept (beta_0): {beta_0}")
print(f"Slope (beta_1): {beta_1}")

# Define the regression model (y = beta_0 + beta_1 * x)
y_pred = beta_0 + beta_1 * x

# Plotting the data and the regression line
plt.scatter(x, y, color='blue', label='Data points')
plt.plot(x, y_pred, color='red', label=f'Regression line: y = {beta_0:.2f} + {beta_1:.2f}x')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression using Least Squares')
plt.legend()
plt.grid(True)
plt.show()

# Example prediction for x = 6
x_new = 6
y_new = beta_0 + beta_1 * x_new
print(f"Prediction for x = {x_new}: y = {y_new}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Given data
x1 = np.array([2, 4, 6])
x2 = np.array([3, 6, 9])
y = np.array([5, 10, 15])

# Create a matrix X where the first column is a column of ones (for the intercept),
# the second column is x1, and the third column is x2.
X = np.column_stack((np.ones(len(x1)), x1, x2))

# Use the normal equation to calculate the coefficients (beta)
# beta = (X.T * X)^(-1) * X.T * y
beta = np.linalg.inv(X.T @ X) @ (X.T @ y)

# Extract the coefficients
beta_0, beta_1, beta_2 = beta

# Print the coefficients
print(f"Intercept (beta_0): {beta_0}")
print(f"Coefficient for x1 (beta_1): {beta_1}")
print(f"Coefficient for x2 (beta_2): {beta_2}")

# Define the regression model
y_pred = X @ beta  # Predicted y values

# Plotting the actual vs predicted values
plt.scatter(range(len(y)), y, color='blue', label='Actual data (y)')
plt.plot(range(len(y)), y_pred, color='red', label='Predicted values (y)')
plt.xlabel('Index')
plt.ylabel('y')
plt.title('Multiple Linear Regression: Actual vs Predicted')
plt.legend()
plt.grid(True)
plt.show()

# Example prediction for new values of x1 = 7, x2 = 10
x1_new = 7
x2_new = 10
y_new = beta_0 + beta_1 * x1_new + beta_2 * x2_new
print(f"Prediction for x1 = {x1_new}, x2 = {x2_new}: y = {y_new}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Given data
x1 = np.array([2, 4, 6])
x2 = np.array([3, 6, 9])
y = np.array([5, 10, 15])

# Step 1: Calculate the means of x1, x2, and y
x1_bar = np.mean(x1)
x2_bar = np.mean(x2)
y_bar = np.mean(y)

# Step 2: Calculate a1 (slope for x1) and a2 (slope for x2)
# Slope for x1
numerator_a1 = np.sum((x1 - x1_bar) * (y - y_bar))
denominator_a1 = np.sum((x1 - x1_bar) ** 2)  # No square for the slope in the numerator
a1 = numerator_a1 / denominator_a1

# Slope for x2
numerator_a2 = np.sum((x2 - x2_bar) * (y - y_bar))
denominator_a2 = np.sum((x2 - x2_bar) ** 2)  # Again, no square for the slope in the numerator
a2 = numerator_a2 / denominator_a2

# Step 3: Calculate a0 (intercept)
a0 = y_bar - a1 * x1_bar - a2 * x2_bar

# Print the coefficients
print(f"Intercept (a0): {a0}")
print(f"Slope for x1 (a1): {a1}")
print(f"Slope for x2 (a2): {a2}")

# Step 4: Predict values based on the model
y_pred = a0 + a1 * x1 + a2 * x2

# Plotting the actual vs predicted values
plt.scatter(range(len(y)), y, color='blue', label='Actual data (y)')
plt.plot(range(len(y)), y_pred, color='red', label=f'Predicted values (y)')
plt.xlabel('Index')
plt.ylabel('y')
plt.title('Multiple Linear Regression: Actual vs Predicted')
plt.legend()
plt.grid(True)
plt.show()

# Example prediction for new values of x1 = 7, x2 = 10
x1_new = 7
x2_new = 10
y_new = a0 + a1 * x1_new + a2 * x2_new
print(f"Prediction for x1 = {x1_new}, x2 = {x2_new}: y = {y_new}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Simple dataset (x, y)
x_vals = np.array([1, 2, 3, 4, 5])
y_vals = np.array([2, 4, 5, 4, 5])

# Initial parameters for the line y = mx + b
m = 0  # Initial slope
b = 0  # Initial intercept

# Learning rate
alpha = 0.1

# Gradient descent for 5 iterations
iterations = 5
history_m = [m]  # To store the history of m values
history_b = [b]  # To store the history of b values

for i in range(iterations):
    # Predictions based on the current m and b
    y_pred = m * x_vals + b
    
    # Compute the gradients for m and b
    dm = (-2 / len(x_vals)) * np.sum(x_vals * (y_vals - y_pred))  # Gradient for m
    db = (-2 / len(x_vals)) * np.sum(y_vals - y_pred)  # Gradient for b
    
    # Update m and b
    m -= alpha * dm
    b -= alpha * db
    
    # Record the history
    history_m.append(m)
    history_b.append(b)

# Final m and b after gradient descent
print(f"Final slope (m): {m}")
print(f"Final intercept (b): {b}")

# Plot the data points and the lines at each iteration
plt.scatter(x_vals, y_vals, color='blue', label='Data points')

# Plot lines for each iteration
x_line = np.linspace(0, 6, 100)
for i in range(len(history_m)):
    plt.plot(x_line, history_m[i] * x_line + history_b[i], label=f"Iteration {i}")

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Gradient Descent for Linear Regression (5 iterations)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a simple dataset (e.g., y = 2x + 1 with some noise)
np.random.seed(0)
X = np.random.rand(100) * 10  # Random values for X between 0 and 10
y = 2 * X + 1 + np.random.randn(100)  # Linear relation with some noise

# Cost function for linear regression (Mean Squared Error)
def cost_function_linear(X, y, m, b):
    n = len(X)
    y_pred = m * X + b  # Predicted y values
    cost = (1 / (2 * n)) * np.sum((y_pred - y) ** 2)  # MSE
    return cost

# Generate values for m (slope) to plot the cost function
m_values = np.linspace(-5, 5, 100)  # Values of m from -5 to 5
b = 1  # Fix b at 1 for simplicity

# Compute the cost for each value of m
cost_values = [cost_function_linear(X, y, m, b) for m in m_values]

# Plot the cost function
plt.figure(figsize=(8, 6))
plt.plot(m_values, cost_values, label="Cost Function (MSE)", color='b')
plt.title("Cost Function for Linear Regression")
plt.xlabel("Slope (m)")
plt.ylabel("Cost (MSE)")
plt.grid(True)
plt.legend()
plt.show()
```
